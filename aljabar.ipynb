```python
bil2 = np.arange(1,5)
```

```python
p1 = np.sum(bil2)
p2 = np.sum(bil2**2)

j1 = p1/p2
j2 = 1/p1
```

```python
print(f'Jawaban 1 hasilnya adlah {j1} \n Jawaban 2 hasilnya adalah {j2}')
```

```python
j1==j2
```

```python
p3 = np.prod(bil2)
p4 = np.prod(bil2**2)

j3 = p3/p4
j4 = 1/p3
```

```python
print(f'Jawaban 1 hasilnya adlah {j3} \n Jawaban 2 hasilnya adalah {j4}')
```

```python
j3==j4
```
