```python
# import Library
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
sns.set()
```

```python
# menentukan rataan 2 kelompk
#grup 1 = tanpa pupuk
#grup 2 = dngan pupuk
rataan1 = 50
rataan2 = 51 # 65

# menentukan standar deviasi(variasi) == sama untuk 2 kelompok
sd = 5

# Ukuran sample dua kelompok
sample1 = 50
sample2 = 40
```

```python
# generate bilangan random

test = np.random.randn(sample1)
test
```

```python
plt.plot(test)
```

```python
sns.displot(test)
```

```python
sns.distplot(test);
```

```python
#generate data
data1 = np.random.randn(sample1)*sd + rataan1
data2 = np.random.randn(sample2)*sd + rataan2

# Menggabungkan sample size (jumlah data)
kelompok_data = [sample1, sample2]

batas = [np.min(np.hstack((data1, data2))), np.max(np.hstack((data1, data2)))]
```

```python
batas
```

```python
# Plot distribusi kedua kelompok (Statistik Deskriptif)
plt.figure(figsize=(7,5))

a = sns.distplot(data1, hist=False, label='Data1')
a = sns.distplot(data2, hist=False, label='Data2')

a.set(xlabel='Data', ylabel='Density')
plt.legend();
```

```python
# melakukan uji statistik(Statistik Inferential)
fig, ax = plt.subplots(1,2, figsize=(7,4), dpi=100)

ax[0].violinplot(data2)
ax[0].plot(1+np.random.randn(sample2)/10, data2, 'o', color='blue')
ax[0].set_ylim(batas)

ax[1].violinplot(data1)
ax[1].plot(1+np.random.randn(sample1)/10, data1, 'o', color='green')
ax[1].set_ylim(batas)

#melaukan uji t-test
t,p = stats.ttest_ind(data1, data2)

# mencetak hasil ttest di judul
sigtxt = ('', 'TIDAK')
plt.title('Dua Kelompok {} berbeda secara signifikan! t ({}) = {}, p={}'.format(sigtxt[int(p>.05)],
                                                                         sum(kelompok_data)- 2,
                                                                         np.round(t, 2),
                                                                         np.round(p, 6)))
plt.show()
```
