```python
import numpy as np
import sympy as sym
from IPython.display import Math
import matplotlib.pyplot as plt
```

```python
baris = np.array([2,3])
kolom = np.array([[1], [4]])
```

```python
print(baris)
print(kolom)
```

```python
display(Math(sym.latex(sym.simplify(baris))))
```

```python
display(Math(sym.latex(sym.simplify(kolom))))
```

```python
plt.plot([0, baris[0]], [0, baris[1]], 'r', label='vektor baris')
plt.plot([0, kolom[0]], [0, kolom[1]], 'b', label='vektor baris')
plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
a = np.array([2,0,-3])
b = np.array([-2,2,4])
```

```python
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

ax.plot([0, a[0]], [0, a[1]], [0, a[2]], 'r', linewidth=3)
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'g', linewidth=3)

ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)
plt.show()
```
