```python

# f(x) = x**2+1
# dif(f(x)) = 2x
# slope = turunan = perubahan f(x)/perubahan x =diff(f(x))/diff(x)


x = np.linspace(-2,5,200)
f = x**2 + 1
dx = np.diff(x)
df = np.diff(f)
```

```python
plt.plot(x,f, label='f')
plt.plot(x[:-1],df/dx, label='df/dx')
plt.grid()
plt.legend()
plt.axis([-2, 5, -10, 20])
plt.show()
```

```python
plt.plot(x[:-1],df/dx)
```

```python
# f(x) = x**2 + 1
# jika x =3, maka f =10
# jika x =1, maka f=2
#slope = peubahan f/perubahan x= (10-2)/(3-1) = 8/2 = 4
```
