```python
#inheritance = pewarisan

#single inheritance
class animal:
  def walk(self):
    print(self.__name__+"walk!")


class cat(animal):
  def walk(self):
    print("Yeayy!", +self.__name__+"walk!")
```

```python
#Multilevel Inheritance
class animal:
  def walk(self):
    print(self.__name__+"walk!")


class carnivore(animal):
  def eat(self):
    print("Yeayy!", +self.__name__+"walk!")

class cat(carnivore):
  pass
```

```python
#Multiple Inheritance
class Ovipar:
   pass
 
class Vivipar():
   pass
 
class Ovovivipar (Ovipar, Vivipar):
   pass
 
class Snake(Ovovivipar):
   pass
```

```python
#hirarki pewarisan
class Animal:
   def walk(self):
       print(self.__name__ + "walk!")
 
class Carnivore(Animal):
   def eat(self):
       print("Yeay! " + self.__name__ + " eat!")
 
class Cat(Carnivore):
   pass
 
class Lion(Carnivore):
   pass
 
class Dog(Carnivore):
   pass
```
